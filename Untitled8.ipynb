##### Student_exam_performance #####

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Student_per.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
semester_mean = df.groupby("Semester")["Marks"].mean()
semester_mean
```

```python
semester_median = df.groupby("Semester")["Marks"].median()
semester_median
```

```python
semester_mode = df.groupby("Semester")["Marks"].agg(lambda x: x.mode().iloc[0])
semester_mode
```

```python
semester_result = df.groupby("Semester")["Result"].value_counts()
semester_result
```

```python
for sem in df["Semester"].unique():
    plt.hist(df[df["Semester"] == sem]["Marks"], bins=8)
    plt.title(f"Marks Distribution - {sem}")
    plt.xlabel("Marks")
    plt.ylabel("Number of Students")
    plt.show()
```

```python
student_sem = df.groupby(
    ["Roll_No", "Semester"]
)["Marks"].mean().unstack()
```

```python
student_sem.plot(kind="bar")
plt.xlabel("Roll Number")
plt.ylabel("Average Marks")
plt.title("Student-wise Semester Performance")
plt.legend(title="Semester")
plt.show()
```

```python
subject_stats = df.groupby("Subject")["Marks"].mean()
subject_stats
```

```python
subject_stats.plot(kind="bar")
plt.xlabel("Subject")
plt.ylabel("Average Marks")
plt.title("Subject-wise Performance Analysis")
plt.show()
```

```python
pass_fail = df["Result"].value_counts()
```

```python
plt.pie(
    pass_fail,
    labels=pass_fail.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Pass vs Fail Analysis")
plt.show()
```

```python
df.boxplot(column="Marks", by="Semester")
plt.title("Semester-wise Marks Spread")
plt.suptitle("")
plt.xlabel("Semester")
plt.ylabel("Marks")
plt.show()
```

```python
semester_mean = df.groupby("Semester")["Marks"].mean()

plt.plot(semester_mean.index, semester_mean.values, marker="o")
plt.xlabel("Semester")
plt.ylabel("Average Marks")
plt.title("Semester-wise Performance Trend")
plt.grid(True)
plt.show()
```

```python
pass_percent = df[df["Result"]=="Pass"].groupby("Semester").size() / df.groupby("Semester").size() * 100

pass_percent.plot(marker="o")
plt.ylabel("Pass Percentage")
plt.xlabel("Semester")
plt.title("Semester-wise Pass Percentage Trend")
plt.grid(True)
plt.show()
```

##### The analysis shows that student performance varies across semesters.
The semester-wise line graph indicates overall improvement in average
marks. Subject-wise analysis reveals strengths and weaknesses, while
passâ€“fail and statistical measures such as mean, median, and mode
provide deeper insight into academic performance. #####
